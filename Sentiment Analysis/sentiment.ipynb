```python
import nltk

# nltk.download()

nltk.download('vader_lexicon')
nltk.download('movie_reviews')
nltk.download('punkt')
from nltk.sentiment.vader import SentimentIntensityAnalyzer as SIA
```

```python
from textblob import TextBlob
```

```python
def sentiment(input_sentence):
  sia = SIA()
  text = str(TextBlob(input_sentence).correct())

  output = sia.polarity_scores(text)
  score = output["compound"]
#   print(text)
  if(score < 0.50  and score > 0 ):
        return "Bot: Ok good to heared that"
  elif(score > 0.50):
    return "Bot: Nice to hear about that"
  elif(score < 0 and score > -0.50):
        return "Bot: oh sorry to hear about this"
  elif(score < -0.50):
    return "Bot: I'm extremely sorry to hear that"
  elif(score == 0):
        return "Bot: can I know more detail about this"
```
